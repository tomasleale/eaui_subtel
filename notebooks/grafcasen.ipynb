# Hogares Conectados (BA + Teléfono)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos
años = [2013, 2015, 2017, 2022, 2024]
hogares = [2444114, 3580043, 4000240, 6506762, 6838113]

# DataFrame
df = pd.DataFrame({
    'Año': años,
    'Hogares conectados': hogares
})

print(df)
```

```python
# Gráfico
plt.figure(figsize=(10, 6))
bars = plt.bar(range(len(df)), df['Hogares conectados'], color='steelblue', edgecolor='black', width=0.6)
plt.title('Hogares Conectados (BA + Teléfono)', fontsize=14, fontweight='bold')
plt.xlabel('Año', fontsize=12)
plt.ylabel('Cantidad de Hogares', fontsize=12)
plt.xticks(range(len(df)), df['Año'].astype(str))
plt.grid(True, alpha=0.3, axis='y')

# Etiquetas de datos con formato español
for i, v in enumerate(df['Hogares conectados']):
    etiqueta = f'{v:,.0f}'.replace(',', '.')
    plt.text(i, v + 100000, etiqueta, ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```
